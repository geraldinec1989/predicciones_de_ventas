```python
df_limpiar=df.dropna(subset=['Item_Weight','Outlet_Size'])
```

```python
# Creen un mínimo de dos visualizaciones de datos para ayudar a otros a comprender las tendencias en los datos (análisis de datos explicativos).
# Puesto que estos gráficos son para la elaboración de informes, asegúrense de que tengan un buen aspecto incluyendo títulos, leyendas, etc.


plt.hist([df_limpiar['Outlet_Type'], df_limpiar['Item_Outlet_Sales']], bins=10, alpha=0.7, label=['Outlet_Type', 'Item_Outlet_Sales'])

# Añadir etiquetas y título
plt.xlabel('Tiendas')
plt.xticks(rotation=45)
plt.ylabel('Total de Ventas')
plt.title('Histograma Tipo de Tienda v/s Total de Ventas')
plt.legend()

# Mostrar el histograma
plt.show()
```

```python
#En este caso se utiliza un boxplot para decifrar por Tamaño de tienda cuales son las que tienen las mayores ventas y el promedio de ventas de cada una.

df.boxplot(column = 'Item_Outlet_Sales', by = 'Outlet_Size');
plt.title('Ventas por tipo de supermercado');
```
