```python
#1.Histograma para ver las distribuciones de las diversas características en el conjunto de datos.

plt.hist(df['Item_Outlet_Sales'], bins=10, color='skyblue', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Histograma de Ventas de Artículos en el Outlet')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.show()
```

```python
#2.Boxplot para ver resúmenes estadísticos de diversas características del conjunto de datos.

df.boxplot(column = 'Item_Outlet_Sales', by = 'Outlet_Type');
plt.title('Ventas por tipo de supermercado');



```

```python
#3.Mapa de calor de correlación entre las características.
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Codin Dojo/sales_predictions_2023 (1).csv')
df[['Item_Weight','Item_Visibility','Item_MRP','Item_Outlet_Sales']].corr()


heatmap = sns.heatmap(df[['Item_Weight','Item_Visibility','Item_MRP','Item_Outlet_Sales']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor')
plt.show()

```
